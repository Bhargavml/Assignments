#### 2Q) A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
####  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Loading Dataset
data = pd.read_csv('LabTAT.csv')
data.head()
```

```python
data.describe()
```

```python
# Generate different plots for normality check
data.hist()

data.plot(kind = 'box')

sm.qqplot(data['Laboratory 1'], line = 'q')
plt.title('Lab 1')

sm.qqplot(data['Laboratory 2'], line = 'q')
plt.title('Lab 2')

sm.qqplot(data['Laboratory 3'], line = 'q')
plt.title('Lab 3')

sm.qqplot(data['Laboratory 4'], line = 'q')
plt.title('Lab 4')

plt.show()
```

```python
# Define the null hypothesis
H0:'There is no difference in the average TAT among the laboratories.'
```

```python
# Define the alternative hypothesis
H1:'There is a difference in the average TAT among the laboratories.'
```

```python
# Perform one-way ANOVA (equivalent to F-test)
f_statistic, f_pvalue = stats.f_oneway(data['Laboratory 1'], data['Laboratory 2'], data['Laboratory 3'], data['Laboratory 4'])

# Print F-test results
print("Test statistic =", f_statistic)
print("p_value =", f_pvalue)
```

```python
# Interpret Results
alpha = 0.05
if f_pvalue < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
